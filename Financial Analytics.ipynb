```python
top_mar_cap = data.sort_values(by='Mar Cap - Crore', ascending=False).head(10)
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.barh(top_mar_cap['Name'], top_mar_cap['Mar Cap - Crore'], color='skyblue')
plt.xlabel('Market Cap (Crore)')
plt.title('Top 10 Companies by Market Cap')
plt.tight_layout()
plt.show()
```

```python
top_sales = data.sort_values(by='Sales Qtr - Crore', ascending=False).head(10)
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 2)
plt.barh(top_sales['Name'], top_sales['Sales Qtr - Crore'], color='salmon')
plt.xlabel('Sales Qtr (Crore)')
plt.title('Top 10 Companies by Sales Qtr')

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Mar Cap - Crore'], data['Sales Qtr - Crore'], color='red')
plt.ylabel('Sales')
plt.xlabel('Market Cap (in billions)')
plt.xticks(rotation=75)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
market_mean = data['Mar Cap - Crore'].mean()
sales_mean = data['Sales Qtr - Crore'].mean()

print(f"Mean of Market Cap: {market_mean}")
print(f"Mean of Sales Qtr: {sales_mean}")
```

```python
plt.figure(figsize=(5, 3))
sns.histplot(data['Mar Cap - Crore'], bins=30, kde=True)
plt.title('Distribution of Market Capital')
plt.xlabel('Market Cap (Crore)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.histplot(data['Sales Qtr - Crore'], bins=30, kde=True)
plt.title('Distribution of Sales Qtr')
plt.xlabel('Sales Qtr (Crore)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Mar Cap - Crore', y='Sales Qtr - Crore', data=data)
plt.title('Market Cap vs. Sales Qtr')
plt.xlabel('Market Cap (Crore)')
plt.ylabel('Sales Qtr (Crore)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.boxplot(y='Mar Cap - Crore', data=data)
plt.title('Boxplot of Market Cap')
plt.ylabel('Market Cap (Crore)')

plt.subplot(1, 2, 2)
sns.boxplot(y='Sales Qtr - Crore', data=data)
plt.title('Boxplot of Sales Qtr')
plt.ylabel('Sales Qtr (Crore)')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Correlations')
plt.show()
```
